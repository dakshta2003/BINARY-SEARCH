```python
def Binarysearch(List1,key):
    Low=0
    high=len(List1)-1
    Found = False
    while Low<=high and not Found:
        mid = (Low+high) //2
        if key == List1[mid]:
           Found = True
        elif key>List1[mid]:
            Low = mid+1
        else:
          high = mid-1
    if Found == True:
        print("roll number is found")
    else:
        print("roll number is not found")
List1=[4,53,7,9,77,22]
List1.sort()
print(List1)
key=int(input("enter the roll number:"))
Binarysearch(List1,key)
```
